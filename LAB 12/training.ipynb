```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Step 1: Load the data
data = pd.read_csv("gender_submission.csv")
```

```python
data['Feature1'] = pd.Series(range(len(data)))  # Example of a sequence feature
data['Feature2'] = pd.Series(len(data) - pd.Series(range(len(data))))  # Reverse sequence as a feature
```

```python
# Define X (features) and y (target)
X = data[['Feature1', 'Feature2']]  # Replace these with real feature columns if available
y = data['Survived']  # Target variable
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train the model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
# Evaluate the model
y_pred = model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.preprocessing import LabelEncoder

# Check if 'Sex' column exists
if 'Sex' in data.columns:
    # Handle missing values (fill with 'unknown' or drop rows with NaN)
    data['Sex'].fillna('unknown', inplace=True)  # Fill NaN with a placeholder (optional)
    
    # Apply Label Encoding
    label_encoder = LabelEncoder()
    data['Sex'] = label_encoder.fit_transform(data['Sex'])
else:
    print("'Sex' column not found in dataset.")
```

```python
from sklearn.preprocessing import LabelEncoder

# Loop through all object columns
for column in data.select_dtypes(include=['object']).columns:
    data[column].fillna('unknown', inplace=True)  # Handle missing values by filling with 'unknown'
    label_encoder = LabelEncoder()
    data[column] = label_encoder.fit_transform(data[column])
```

```python
# Check if 'Embarked' column exists
if 'Embarked' in data.columns:
    # Handle missing values in 'Embarked' column (optional: fill with a placeholder value)
    data['Embarked'].fillna('Unknown', inplace=True)  # Replace NaN with 'Unknown'
    
    # Apply One-Hot Encoding using pd.get_dummies()
    data = pd.get_dummies(data, columns=['Embarked'], drop_first=True)
else:
    print("'Embarked' column not found in the dataset.")
```

```python
from sklearn.preprocessing import LabelEncoder

# Loop through all object (categorical) columns
for column in data.select_dtypes(include=['object']).columns:
    label_encoder = LabelEncoder()
    data[column] = label_encoder.fit_transform(data[column])
```
