```python
#actual lab task  
class TempCtrl:
    def __init__(self, goal_temp):
        self.goal_temp = goal_temp
        self.temp_history = {}

    def set_temp(self, space, curr_temp):
        self.space = space
        self.curr_temp = curr_temp

    def decide(self):
        return "Heater Off" if self.curr_temp > self.goal_temp else "Heater On"

    def check_history(self, space, temp):
        return self.temp_history.get(space + "_action") if space in self.temp_history and self.temp_history[space] == temp else None

    def log_temp(self, space, temp, action):
        self.temp_history[space] = temp
        self.temp_history[space + "_action"] = action

    def load_log(self, filename):
        try:
            with open(filename, "r") as file:
                for line in file:
                    space, temp, action = line.strip().split(",")
                    self.log_temp(space, int(temp), action)
        except FileNotFoundError:
            pass

    def save_log(self, filename):
        with open(filename, "w") as file:
            for space, temp in self.temp_history.items():
                if "_action" not in space:
                    file.write(f"{space},{temp},{self.temp_history[space + '_action']}\n")


def main():
    ctrl = TempCtrl(22)
    ctrl.load_log("temp_log.txt")

    while True:
        print("\nOptions:")
        print("1. Enter space and temperature")
        print("2. Display temperature log")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            space = input("Enter space name: ")
            temp = int(input("Enter temperature: "))
            existing_log = ctrl.check_history(space, temp)
            if existing_log:
                print(f"{space}: {temp} ==> {existing_log}")
            else:
                ctrl.set_temp(space, temp)
                action = ctrl.decide()
                ctrl.log_temp(space, temp, action)
                print(f"{space}: {temp} ==> {action}")
                ctrl.save_log("temp_log.txt")
        elif choice == "2":
            print("\nTemperature Log:")
            with open("temp_log.txt", "r") as file:
                for line in file:
                    space, temp, action = line.strip().split(",")
                    print(f"{space}: {temp} ==> {action}")
        elif choice == "3":
            break
        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```

```python
#lab3 code in class not a task
class SimpleReflexAgent:
    def __init__(self, temp):
        self.desired_temp = temp
        self.current_temp = None

    def percept(self, temp):
        self.current_temp = temp

    def act(self):
        if self.current_temp > self.desired_temp:
            return "Turn off the heater"
        else:
            return "Turn on the heater"


# Create an agent with desired temperature 22
agent = SimpleReflexAgent(22)

# Test the agent with temperature 16
agent.percept(16)
print("Test with temperature 16:")
print(agent.act())

# Test the agent with multiple rooms
rooms = {
    "Living room": 28,
    "Bedroom": 18,
    "Kitchen": 32
}

print("\nRoom Temperature Control:")
for room, temp in rooms.items():
    agent.percept(temp)
    print(f"{room}: {temp} ==> {agent.act()}")

```
